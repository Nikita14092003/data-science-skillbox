### Вопросы по лекциям.

Как поймать вообще все ошибки, которые могут произойти?

**Ответ:**
try:
except: # без указания конкретного класса ошибки

Сколько раз подряд можно указывать except?

**Ответ:**
Сколько угодно, указывая разные классы ошибок (один и тот же класс можно указать, интерпретатор не будет ругаться, но ветки, кроме первой, не выполнятся никогда)

Вы хотите с помощью print вывести название ошибки в консоль, как это сделать?

**Ответ:**
Не совсем поняла, что именно в вопросе понимается под названием, но можно вывести так:
try:
except <класс исключения> as exc: # или любое другое имя переменной
    print(f'произошла ошибка: {exc}')

Вы хотите с помощью print вывести параметры ошибки в консоль, как это сделать?

**Ответ:**
try:
except <класс исключения> as exc: # или любое другое имя переменной
    print(f'параметры ошибки: {exc.args}')

Что такое DeprecationWarning?

**Ответ:**
Тип варнинга ("ошибка", которая не приводит к остановке программы), который указывает на то, что используется "неактуальная" часть кода (функция, например, или константа), её поддержка может быть прекращена в следующих версиях и обычно есть более актуальный для библиотеки способ сделать то же самое.

### Задание 1.

Есть файл с протоколом регистраций пользователей на сайте (registrations.txt).<br>
Каждая строка содержит информацию о имени, электронной почте и возрасте человека. <br><br>


Надо проверить данные из файла, для каждой строки:
 - присутсвуют все три поля
 - поле имени содержит только буквы
 - поле email содержит @ и .
 - поле возраст является числом от 10 до 99<br>

В результате проверки нужно сформировать два файла
 - registrations_good.log для правильных данных, записывать строки как есть
- registrations_bad.log для ошибочных, записывать строку и вид ошибки.<br>

Для валидации строки данных написать метод, который может выкидывать исключения:
 - НЕ присутсвуют все три поля: ValueError
 - поле имени содержит НЕ только буквы: NotNameError (кастомное исключение)
 - поле email НЕ содержит @ и .(точку): NotEmailError (кастомное исключение)
 - поле возраст НЕ является числом от 10 до 99: ValueError
Вызов метода обернуть в try-except.

```python
class NotNameError(Exception):
    pass


class NotEmailError(Exception):
    pass


def validate_reg_line(val_line):
    val_line = val_line.strip()
    params = val_line.split(sep=' ')
    
    if len(params) < 3:  # предполагаем, что в строке могут быть ещё данные, но минимум три
        raise ValueError('В строке данных регистрации должны присутствовать минимум три поля')

    name, email, age = params[0], params[1], params[2]  # не пишу ...= params из предположения, что полей м.б. больше
    if not name.isalpha():
        raise NotNameError('Поле Имя не валидно. В поле имени содержатся не только буквы')

    email_parts = email.split(sep='@')
    if len(email_parts) != 2:
        raise NotEmailError("Поле e-mail не валидно. Поле должно содержать ровно 1 знак '@'")
    elif len(email_parts[1].split('.')) != 2:
        raise NotEmailError("Поле e-mail не валидно. Поле должно содержать ровно 1 знак '.' после '@'")

    # здесь поправила, нельзя было двойное сравнение указывать в данном случае, 
    # оно же работает как и, а не как или и всегда ложно было
    if not age.isnumeric() or int(age) > 99 or int(age) < 10:
        raise ValueError('Поле возраст не валидно. Поле должно содержать число от 10 до 99')


def write_line_to_bad_log(file, line, exc):
    file.write(f"Строка: '{line[:-1]}'. Ошибка: '{exc}' \n")
    print(f"Строка: '{line[:-1]}'. Ошибка: '{exc}'")


# здесь я уверена, что файл есть, это не данные пользователя, поэтому не обрабатываю возможное отсутствие файла
with open(file='registrations_.txt', mode='r', encoding='utf-8-sig') as reg_file:

    good_log = open('registrations_good.log', 'w', encoding='utf-8')
    bad_log = open('registrations_bad.log', 'w', encoding='utf-8')

    for line in reg_file:
        try:
            validate_reg_line(line)
            good_log.write(line)
        except ValueError as exc:
            write_line_to_bad_log(bad_log, line, exc)
        except NotNameError as exc:
            write_line_to_bad_log(bad_log, line, exc)
        except NotEmailError as exc:
            write_line_to_bad_log(bad_log, line, exc)

    good_log.close()
    bad_log.close()
```
