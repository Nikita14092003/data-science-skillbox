### 5.
Постройте модель на данных *x_train_1, y_train_1* с гиперпараметром n_neighbors, пробегающим значения от 1 до 20 включительно, и укажите значения n_neighbors, которым соответствует наиболее высокий результат функции accuracy_score().

*Примечание: можно воспользоваться циклом for, чтобы не прописывать вручную все 20 вариаций модели.*

```python
accuracy = []

for n_neighbors in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn_model = knn.fit(x_train_1, y_train_1)
    knn_predictions = knn.predict(x_test_1)
    accuracy.append(accuracy_score(y_test_1, knn_predictions))

accuracy = np.array(accuracy)
np.argmax(accuracy)
```

```python
accuracy_indecies_sort = accuracy.argsort()
max_accuracy = accuracy[accuracy_indecies_sort[-1]]
max_accuracy
```

```python
best_n = []
for i in accuracy_indecies_sort[::-1]:
    if accuracy[i] == max_accuracy:
        best_n.append(i + 1) # количество соседей равно индексу + 1, соседей мы считаем с 1, индексы - с 0
    else:
        break # как только не равно, значит попали на индекс меньше и индексы максимальных элементов закончились
        
print('Значения параметра Количество соседей, которым соответствует наилучшая точность на наших тестовых данных')
best_n
```
